## Exploring USAID Public-Private Partnership Investment in the Global Development Alliance

### Introduction

#### Background on PPPs
Public-private partnerships (PPPs)- a nebulous term most commonly conceptualized as a long-term contract between the public sector and private enterprise to execute or operate a project- are a growing feature of global public administration and international development. The United Stated Agency for International Development (USAID) formalized its PPP development model as the Global Development Alliance (GDA) in 2001.

While acknowledging that PPPs are diverse and continually evolving, the model’s inherent tension between private interests and public accountability make it uniquely exploitable and the subject of debate about the commodification of public goods with private finance, financialization of public infrastructure, delivery on promises of private sector efficiency, and fiscal risk. The promise of PPPs is not only in taking advantage of private sector efficiency or, often dubious, claims of public cost-saving, but in the assumption that PPPs provide “better value” for money than the public sector alone can achieve. That the proliferation of PPPs has grown in tandem with the rise of neoliberalism, New Public Management, and the perceived weakening of the state in the face of global capitalism is hard to ignore.


#### Rationale for this Project
For the purposes of this project, it will be interesting to explore what sorts of countries/regimes are recieving PPP investment and discern what governmental and institutional characterstics of countries are associated with this investment. US taxpayers, who fund USAID's part in the GDA, should have some notion (and by extension, tool for accountability) about where and how their money is being used. Furthermore, we can learn something about USAID's motivations and priorities if we know where its investment is being directed. Is it being channeled to countries/regimes that reflect the values of the Department of State- namely democratic governance, combatting corruption, and transparent development assistance? Or to countries with weaker institutions, lower levels of democracy, more internal conflict, etc. in an effort to bolster and promote the sort of developments USAID wishes to see?

I chose eight key indicators to capture quality of governance and regime characteristics. These are elaborated on in the Data Description section.

#### Summary of Findings

As a proportion of their investment, the US Government and Non-US Government partners are investing at the same levels in all five regions. As for sectors, Health, Agriculture & Food Security, and Environment are the most numerous but the volume of programming does not translate to corresponding levels of investment. Indeed, the Education sector attracts the most investment overall.

There are very low levels of correlation between amount of investment and index scores on all eight indices, and only five of the eight are statistically significant. Single variate regression analysis reveals that countries with higher levels of corruption, worse public services, more socio-ethnic-religioius conflict, and more impact from terrorism are also associated with large increases in investment. Countries with better state of democracy could expect a decrease in US Government investment. Multiple regression models reveal that Non US Government partners invest more in countries with higher levels of socio-ethnic-religious tension while the US Government tends to invest more in countries suffering greater impacts of global terrorism. Non US Governments partners also tend to invest more in regimes with higher levels of political terror (violations of human rights) than the US Government. Neither multiple regression model is statistically significant however.

  #### Explanation of the Eight Indices (Columns 13-20):

    - **Bayesian Corruption** is a score between 0 and 100 where an increase in the index corresponds to an increased level of corruption (i.e., the lower the score the better). It is a composite index maintained by Sherppa Ghent University of the perceived overall level of corruption, with corruption defined as the "abuse of public power for private gain." 0 = there is no corruption in the country and 100 = corruption is as bad as it gets.
    - **State Fragility** is a score between 0 and 25 that provides annual state fragility, effectiveness, and legitimacy indices and is produced by the Center for Systemic Peace. An increase in the index corresponds to higher state effectiveness and legitimacy (i.e., the higher the score, the better). A country's fragility is closely associated with its "state capacity to manage conflict; make and implement public policy; and deliver essential services and its systemic resilience in maintaining system coherence, cohesion, and quality of life; responding effectively to challenges and crises and sustaining progressive development."  
    - **Political Terror** is a score between 1 and 5 which measures "violations of basic human rights carried out by states or their agents" based on yearly country reports from Amnesty International. The lower the score, the better. An increase in the index corresponds to higher political terror where 1 = “Countries under a secure rule of law, people are not imprisoned for their view, and torture is rare or exceptional. Political murders are extremely rare” and 5 = “Terror has expanded to the whole population. The leaders of these societies place no limits on the means or thoroughness with which they pursue personal or ideological goals.”
    - **Democracy Status** is a score from 1-10 within the Bertelsmann Stiftung Transformation Index (BTI) which "analyzes and evaluates the quality of democracy, a market economy and political management in 137 developing and transition countries." The state of democracy is measured with five criteria including stateness, political participation, rule of law, stability of the democratic institutions, and political and social integration. An increase in the index corresponds to a better state of democracy (i.e., a higher score is better).
    - **Public Services** is an indicator between 0-10 that represents “pressures and measures related to policing, criminality, education provision, literacy, water and sanitation, infrastructure, quality healthcare, telephony, internet access, energy reliability, and roads.” 0 = lowest intensity (most stable) and 10 = highest intensity (least stable), therefore the lower the score the better. It is produced by The Fund for Peace through its Conflict Assessment Software Tool which uses social science methodology to triangulate and critically review scores.
    - **E-Government** is a composite measure of three dimensions of an effective government and its ability to provide services and community including provision of online services, telecommunications infrastructure, and human capital. Score range from 0-1 (0-100%), with higher scores indicating greater e-government capability (i.e., the higher the score, the better). It is compiled by the United Nations Department of Social and Economic Affairs.
    - **Conflict Intensity** is a score from 1-10 which measures how serious social, ethnic, and religious conflicts are in a given country. 1 = “There are no violent incidents based on social, ethnic or religious differences” and 10 = “There is civil war or a widespread violent conflict based on social, ethnic or religious differences. It is part of the BTI.
    - **Global Terrorism** is a score from 1-10 which measure the direct and indirect impact of terrorism on a given country including “lives lost, injuries, property damage, and psychological after-effects.” A higher score indicates greater impact of terrorism (so a lower score is better). It is produced by the Institute for Economics and Peace which aggregates data from the Global Terrorism Database into a composite score- it includes domestic and international terrorist incidents.

Countries with higher Public Services scores (or worse public services), higher Conflict Intensity (more socio-ethnic-religious conflict), and higher Global Terrorism (more impact from terrorism), are also associated with large increases in investment. So far that doesn't seem like a positive development for US taxpayers. But increases in the State Fragility score are also associated with an increase in USG investment- and  in this case, the higher the score the better (i.e., a more stable and legitimate state). Worryingly, however, an increase in Politcal Terror is associated with a very large 24% increase in USG funding. Recall that a higher Political Terror score translates to increased severity of human rights violations among a country's people. So too for E-Government, where having an effective, transparent e-government apparatus is associated with a large decrease in funding. Interestingly, a country with a better state of democracy (higher Democracy Status score) could expect a 7.5% decrease in USG investment, suggesting that the US Government is not investing in countries that reflect its values but rather those with less robust states of democracy. We can extrapolate from this analysis that perhaps states with weaker regimes and poorer quality of governance attract more PPP investment due to their inherent needs/perceived weakness.

 
But how significant are these values?

#### Testing Significance

Echoing the results of the model prediction and residual plot above, we can see a very low R-squared of 0.084 for Bayesian Corruption. The proportion of variation in USG Investment that can be attributed to Bayesian Corruption is very low in the context of our model, explaining only about 8.4% of the variation in the dependent variable, USG Investment, around its mean.

The p-value for each independent variable is testing the null hypothesis that the variable has no correlation with the dependent variable, USG Investment.
From the table below, we can see that Bayesian Corruption is statistically significant at the 0.05 significance level (0.005 < 0.05), allowing us to reject the null hypothesis. This sets up an interesting scenario where the independent variable is significant, yet the r-squared value is low. This may indicate that the independent variable, Bayesian Corruption, is correlated with USG Investment, but does not explain much of the variability in USG Investment. 

```python
corruption_model = smf.ols(
    "log_usg_investment ~ bayesian_corruption", data=df_log)
corruption_result = corruption_model.fit()
display(corruption_result.summary())
```

State Fragility is also statistically significant at the 0.05 significance level (0.010 < 0.05), allowing us to reject the null hypothesis. As is Public Services (0.003 < 0.05), Conflict Intensity (0.014 < 0.05), and Global Terrorism (0.028 < 0.05). But Political Terror is not statistically significant at the 0.05 significance level (0.064 > 0.05), so we fail to reject the null hypothesis. Democracy Status is not statistically significant at the 0.05 significance level (0.289 > 0.05), leading us to fail to reject the null hypothesis as well. E-Government (0.074 > 0.05) is also not statistically significant.

```python
fragility_model = smf.ols("log_usg_investment ~ state_fragility", data=df_log)
fragility_result = fragility_model.fit()
display(fragility_result.summary())

terror_model = smf.ols("log_usg_investment ~ political_terror", data=df_log)
terror_result = terror_model.fit()
display(terror_result.summary())

demo_model = smf.ols("log_usg_investment ~ democracy_status", data=df_log)
demo_result = demo_model.fit()
display(demo_result.summary())

services_model = smf.ols("log_usg_investment ~ public_services", data=df_log)
services_result = services_model.fit()
display(services_result.summary())

egov_model = smf.ols("log_usg_investment ~ e_government", data=df_log)
egov_result = egov_model.fit()
display(egov_result.summary())

conflict_model = smf.ols(
    "log_usg_investment ~ conflict_intensity", data=df_log)
conflict_result = conflict_model.fit()
display(conflict_result.summary())

terrorism_model = smf.ols("log_usg_investment ~ global_terrorism", data=df_log)
terrorism_result = terrorism_model.fit()
display(terrorism_result.summary())
```

So five out of the eight independent variables (Bayesian Corruption, State Fragility, Public Services, Conflict Intensity, and Global Terrorism) are the statistically significant. There are three that not statistically significant (Democracy Status, Political Terror, and E-Government). Even though the r-squared values are very low, interpretations of the coefficients of significant variables can still hold, allowing conclusions to be drawn about the relationships between these pairs of variables.

#### Pre Registrated Analysis 2: Comparing USG and Non-USG Funding

```python
# Multiple Regression model for usg and non usg

usg_multi_model = LinearRegression()
usg_multi_model.fit(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'e_government', 'conflict_intensity', 'global_terrorism']], df_log['log_usg_investment'])
usg_multi_score = usg_multi_model.score(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'e_government', 'conflict_intensity', 'global_terrorism']], df_log['log_usg_investment'])
print('USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism: \n',
      usg_multi_model.coef_.round(4), '\n R squared:', usg_multi_score.round(3))

print('--------------------------------------------------------------------------')

nonusg_multi_model = LinearRegression()
nonusg_multi_model.fit(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'e_government', 'conflict_intensity', 'global_terrorism']], df_log['log_non_usg_investment'], )
nonusg_multi_score = nonusg_multi_model.score(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'e_government', 'conflict_intensity', 'global_terrorism']], df_log['log_non_usg_investment'])
print('Non-USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism:\n',
      nonusg_multi_model.coef_.round(4), '\n R squared:', nonusg_multi_score.round(3))
```

```python
# Exponentiated Coefficients to look at % change

usg_pc = np.expm1(usg_multi_model.coef_)*100
print('USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism: \n', usg_pc.round(2))
print('---------------------------------------------------------------------------')
non_usg_pc = np.expm1(nonusg_multi_model.coef_)*100
print('Non USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism: \n', non_usg_pc.round(2))
```

Comparing USG and Non USG Investment multivariate models, a few things stand out. First increases in State Fragility and Political Terror have very different affects on the dependent variable. For USG Investment, this relationship is positive and for Non USG Investment it's negative. Bayesian Corruption and Public Services seem to have similar effects in both models. A 1 unit increase in Conflict Intensity is associated with a much larger % increase in Non USG Investment than USG Investment (13% vs 4%). The opposite is true for Global Terrorism (.88% vs 3.2%). This could suggest that Non USG partners tend to invest/are more willing to invest in countries with higher levels of socio-ethnic-religious conflict while the US Government is more likely/willing to invest in places that have suffered greater effects of both domestic and international terrorism.

One number that seems very strange is E-Government, where a 1 unit increase in E-Government is associated with a massive 522% increase in USG Investment- likewise for the Non USG model where it is associated with a 623% increase. This suggests there is something seriously wrong, likely excessive multi-colinearity. I suspect the problem is that E-Government and Public Services are highly correlated features. Removing one of the suspected highly correlated feature results in more sane coefficients, as seen below.

```python
# multiple regression model without highly correlated feature

usg_multi_model = LinearRegression()
usg_multi_model.fit(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'conflict_intensity', 'global_terrorism']], df_log['log_usg_investment'])
usg_multi_score = usg_multi_model.score(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'conflict_intensity', 'global_terrorism']], df_log['log_usg_investment'])
print('USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism: \n',
      usg_multi_model.coef_.round(4), '\n R squared:', usg_multi_score.round(3))

print('--------------------------------------------------------------------------')

nonusg_multi_model = LinearRegression()
nonusg_multi_model.fit(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'conflict_intensity', 'global_terrorism']], df_log['log_non_usg_investment'], )
nonusg_multi_score = nonusg_multi_model.score(
    df_log[["bayesian_corruption",
            "state_fragility", 'political_terror', 'democracy_status', 'public_services', 'conflict_intensity', 'global_terrorism']], df_log['log_non_usg_investment'])
print('Non-USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, E-Government, Conflict Intensity, Global Terrorism:\n',
      nonusg_multi_model.coef_.round(4), '\n R squared:', nonusg_multi_score.round(3))
```

```python
# Exponentiated Coefficients to look at % change

usg_pc = np.expm1(usg_multi_model.coef_)*100
print('USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, Conflict Intensity, Global Terrorism: \n', usg_pc.round(2))
print('---------------------------------------------------------------------------')
non_usg_pc = np.expm1(nonusg_multi_model.coef_)*100
print('Non USG Investment | Corruption, State Fragility, Political Terror, Democracy Status, Public Services, Conflict Intensity, Global Terrorism: \n', non_usg_pc.round(2))
```

Now we can see the the affects of the independent variables on USG and Non USG Investment respectively are more aligned. For instance, a 1 unit increase in State Fragility is now associated with a negative percent change in both USG Investment and Non USG investment (-0.53% and -2.66% respectively). Conclusions about Conflict Intensity and Global Terrorism still hold, but the much larger coefficient for Political Terror may suggest that Non USG partners tend to invest more or are more willing to invest more in regimes with higher levels of political terror. Unfortunately, however, none of the variables in either model are significant at the 0.05 level- as demonstrated by the high p-values in the tables below. Overall the models do not perform very well and is it possible that the normalized data violate some assumptions of the OLS model.

```python
usg_ols_model = smf.ols(
    "log_usg_investment ~ bayesian_corruption + state_fragility + political_terror + democracy_status + public_services + conflict_intensity + global_terrorism", data=df_log)
usg_ols_result = usg_ols_model.fit()
display(usg_ols_result.summary())

nonusg_ols_model = smf.ols(
    "log_non_usg_investment ~ bayesian_corruption + state_fragility + political_terror + democracy_status + public_services + conflict_intensity + global_terrorism", data=df_log)
nonusg_ols_result = nonusg_ols_model.fit()
display(nonusg_ols_result.summary())
```

Ultimately, some interesting conclusions, like those above, can be drawn from the multiple regression models about the investment behavior of US Government and Non US Government partners but the results are not statisfically significant.

### (Additional) Interpretation and Conclusions

Though Non US Government partners are investing more across the board, both the US Government and Non-US Government partners are investing the most in Africa, followed closely by the Middle East and North Africa (MENA) region. Some of this can be explained by the types of programs within these regions. In terms of number of programs the Health sector, Agriculture & Food Security sector, and Environment sector are the most numerous. Breaking down sector programs by region, Health sector programming is indeed most common in Africa. Yet, curiously, the region with the second largest level of investment, MENA, has only nine Health programs and overall, the fewest programs in total. Indeed, the volume of programming does not translate to corresponding levels of investment, which which sheds more light on why MENA is so close to Africa in terms of investment recieved, despite having far fewer programs. In fact, the Education sector attracts the most investment overall.

Unfortunately, the linear models do not perform very well, both single variate and multiple regression. Only five of the eight linear regressions were statistically significant, and none of the multiple regressions were statistically significant. There are very low levels of correlation between amount of investment and index scores on all eight indices. Single variate regression analysis reveals that countries with higher levels of corruption, worse public services, more socio-ethnic-religioius conflict, and more impact from terrorism are also associated with large increases in investment. Interestingly, a country with a better state of democracy could expect a decrease in US Government investment, suggesting that the US Government is not investing in countries that reflect its values but rather those with less robust states of democracy. We can extrapolate that perhaps states with weaker regimes and poorer quality of governance attract more PPP investment due to their inherent needs/percieved weaknesses.

There is not a lot of difference US Government and Non-USG Government investment, though there are some key differences. Non US Government partners appear to invest more in countries with higher levels of socio-ethnic-religious tension while the US Government tends to invest more in countries suffering greater impacts of global terrorism. Finally, in a key diversion, Non US Governments partners tend to invest more in regimes with higher levels of political terror (violations of human rights) than the US Government.

### Limitations

One of the limitations is in the datasets themselves. According to the USAID website, "the data has been collected according to different mechanisms, definitions, and timeframes from 2001-2015, and therefore cannot be considered to be comprehensive of all PPPs during this period." Furthermore, it does not break down investment by individual sector or by individual resource/implementing partners, making it difficult to parse exactly how funds have been distributed.

Two things also about government data in international development; 1) classified and sensitive information is always scrubbed before public release, potentitally skewing the analysis and 2) corruption, bribery ('cost of doing business'), and kickbacks (from the private and well as public and NGO sectors) can all contribute to the misapproriation of investment and inaccurate reporting.

As outlined throughout the report, the indices were also a limitation. Given that they are all composites of scores from different datasets collected by different organizations, it's very difficult to control for collinearity when choosing them. While indices are a useful way of quantifying socio-political phenomena that are difficult to quantiy (like democracy), the wildly different, and sometimes proprietary, metholody of creating them makes it difficult to truly understand the potential harms and human impact in data collection. Furthermore, there may not have been a very linear relationship between the dependent and independent variables. It also appears that at least one index, Political Terror, may be ordinal rather than continuous.
